```python
import yfinance as yf #Yahoo 주식 데이터 크롤링 모델
import pandas as pd #CSV 파일 인덱싱
import matplotlib.ticker as mticker

ticker_listdff = ['NVDA', 'GOOG', 'Gold'] # Define the ticker for the stock you want to get data for
ticker_list = ['NVDA', 'GOOG', 'ORCL', 'MSFT', 'AMZN']
```

```python
for ticker in ticker_list:
    data = yf.download(ticker, period="5y", interval="1wk")#
    data.to_csv(ticker+".csv")
# Download 3 months of historical data using the daily time frame
```
